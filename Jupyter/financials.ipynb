```python
# import libraries
import pandas as pd;
```

```python
# read CSVs
financials_2018 = pd.read_csv('../data/2018_Financial_Data.csv');
```

```python
# print all industries available
financials_2018.Sector.unique()
```

```python
# clean up data
# we only want companies from the consumer packaged goods industry, "Consumer Cyclical".
# we can consider "Consumer Defensive" as well?
consumer_2018 = financials_2018.loc[financials_2018['Sector'] == "Consumer Cyclical"];
```

```python
#financials_2018_cols = [c for c in financials_2018.columns if c.lower()[-5:] != 'ratio']
#financials_2018_cols = [c for c in financials_2018_cols if c.lower()[-6:] != 'ratios']
financials_2018_cols = [c for c in financials_2018.columns if c.lower()[-6:] != 'margin']
financials_2018_cols = [c for c in financials_2018_cols if c.lower()[-8:] != 'turnover']
financials_2018_cols = [c for c in financials_2018_cols if c.lower()[-6:] != 'growth']
financials_2018_cols = [c for c in financials_2018_cols if c.lower()[-18:] != 'growth (per share)']
financials_2018_cols = [c for c in financials_2018_cols if c.lower()[-9:] != 'per share']
financials_2018_cols = [c for c in financials_2018_cols if c.lower()[-8:] != 'pershare']
```

```python
financials_2018 = financials_2018[[financials_2018_cols.tolist()]]
```

```python
financials_2018 = financials_2018.drop(['cashConversionCycle', 'operatingCycle'], axis=1)
```
