```python
import numpy  as np
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.cluster import AffinityPropagation
```

```python
df = pd.read_csv("../processed_data/2018_Data_Ratios.csv")

# convert 'Name' column into row indices
df.set_index('Name', inplace=True)

# remove ticker column
df_ticker = df['Ticker']
df = df.drop(columns=['Ticker'])
df = df.drop(columns=['Intangibles to Total Assets'])

# display
display(df)
```

```python
random_state = 0
af = AffinityPropagation(preference=-100).fit(df)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
labels_pred = af.predict(df)
```

```python
n_clusters_ = len(cluster_centers_indices)
print('Estimated number of clusters: %d' % n_clusters_)
```

```python
'''first couple labels aren't really relevant'''
print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_pred, labels))
print("Completeness: %0.3f" % metrics.completeness_score(labels_pred, labels))
print("V-measure: %0.3f" % metrics.v_measure_score(labels_pred, labels))
print("Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(labels_pred, labels))
print("Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(labels_pred, labels))
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(df, labels, metric='sqeuclidean'))
```
