### Yahoo Finance API
##### Used to convert ticker names to company names

```python
def getCompanyName(ticker):
    url = "http://d.yimg.com/autoc.finance.yahoo.com/autoc?query={}&region=1&lang=en".format(ticker)

    result = requests.get(url).json()

    for x in result['ResultSet']['Result']:
        if x['symbol'] == ticker:
            return x['name']
```

##### Calls the Yahoo API. Only run if necessary. Takes a minute or two to get all the data from Yahoo servers.

```python
consumer_2018_accounts_final = consumer_2018_accounts_scaled
consumer_2018_accounts_final['Name'] = consumer_2018_accounts_final.apply(lambda row : getCompanyName(row.name), axis = 1)
```

##### Removes Tickers that are not found on Yahoo Finance

```python
consumer_2018_accounts_final = consumer_2018_accounts_final[consumer_2018_accounts_final['Name'].isnull() == False]
```
